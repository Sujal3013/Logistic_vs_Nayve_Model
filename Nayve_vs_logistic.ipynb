# IMPORTING REQUIRED LIBRARIES


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("IRIS.csv")
```

```python
data.head()
```

```python
sns.pairplot(data,hue='species')
```

```python
dataset=data.values
```

```python
X=dataset[:,:4]
Y=dataset[:,4]
```

```python
X[0:4,:]
```

```python
Y[0:4]
```

# Importing scikitlearn & its classes for Model


```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#for 80% train data and 20 % test data
X_train_80,x_test_20,Y_train_80,y_test_20=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
Nayve_model=GaussianNB()
```

```python
Logistic_model=LogisticRegression()
```

```python
Nayve_model.fit(X_train_80,Y_train_80)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain_80 = sc_x.fit_transform(X_train_80) 
xtest_20 = sc_x.transform(x_test_20)
```

```python
Logistic_model.fit(xtrain_80,Y_train_80)
```

```python
Nayve_prediction_80_20=Nayve_model.predict(x_test_20)
```

```python
Logistic_prediction_80_20=Logistic_model.predict(xtest_20)
```

```python
print("Nayve Model Accuracy")
print(accuracy_score(Nayve_prediction_80_20,y_test_20))
```

```python
print("Logistic Model Accuracy")
print(accuracy_score(Logistic_prediction_80_20,y_test_20))
```

```python
Nayve_confusion_matrix_80_20 = metrics.confusion_matrix(y_test_20, Nayve_prediction_80_20)

nayvecm_display_80_20 = metrics.ConfusionMatrixDisplay(confusion_matrix = Nayve_confusion_matrix_80_20,display_labels = ['Setosa','Versicolor','Virgina'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(Nayve_prediction_80_20,y_test_20))
```

```python
Logistic_confusion_matrix_80_20 = metrics.confusion_matrix(y_test_20, Logistic_prediction_80_20)

logisticcm_display_80_20 = metrics.ConfusionMatrixDisplay(confusion_matrix = Logistic_confusion_matrix_80_20,display_labels = ['Setosa','Versicolor','Virgina'])
```

```python
print(classification_report(Logistic_prediction_80_20,y_test_20))
```
